```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
#Parametric Equation r(t)
fig = plt.figure()
ax = plt.axes(projection='3d') 

#parameter t
t = np.linspace(0.0,100.0,1000)

#parametric equations x,y,z
x = np.cos(t)
y = np.sin(t)
z = t

ax.plot3D(x,y,z)  
plt.xlim(-2,2)
plt.ylim(-2,2)
ax.set_zlim(0,100)

ax.view_init(30, 60) #first is rotation around screen horizontal ^----, second is rotation around screen vertial >|>
plt.show()
```

```python
def f(x,y):
    return (4*np.power(x,2)) + (np.power(y,2))
```

```python
#Quadric Surface z = f(x,y)
fig = plt.figure()
ax = plt.axes(projection='3d') 

t = np.linspace(-3,3,1000)
X,Y = np.meshgrid(t,t)
Z = f(X,Y)
ax.plot_surface(X,Y,Z)

a = 10
plt.xlim(-a,a)
plt.ylim(-a,a)
ax.set_zlim(-1,50)

ax.view_init(30, 30)
plt.show()
```
